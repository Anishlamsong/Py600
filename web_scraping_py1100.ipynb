```python
import requests # extracts html from website
```

```python
url = "https://quotes.toscrape.com/"
```

```python
html = requests.get(url).content
```

```python
# html # we want to convert this string to tree
```

```python
# we need BeautifulSoup
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
quotes_list = soup.find_all(class_='text')
quotes_list[0]
```

```python
quotes_list[1].text
```

```python
quotes = [item.text for item in quotes_list]
quotes
```

```python
author_list = soup.find_all(class_ = 'author')
author = [item.text for item in author_list]
```

```python
data = {
    "quotes": quotes, "author": author
}
```

```python
import pandas as pd
```

```python
# table lai DataFrame ma represent garxa pandas le
```

```python
df = pd.DataFrame(data)
df.head() # first 5 rows
```

```python
df.tail() # last 5 rows
```

```python
df.to_csv('quote.csv', index=False)
```

```python
import sqlite3
con = sqlite3.connect('quo.db')
cur = con.cursor()
try:
    cur.execute("create table quotes (quote varchar(100), author varchar(50))")
except Exception as e:
    print(e)
```

```python
# insert
```

```python
cur.executemany("insert into quotes(quote, author) values (?, ?)", df.to_records(index=False))
```
